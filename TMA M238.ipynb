```python
import folium

# Example data: city names with corresponding latitude and longitude
cities = {
    "City1": {"lat": 34.0522, "lon": -118.2437},
    "City2": {"lat": 40.7128, "lon": -74.0060},
    "City3": {"lat": 37.7749, "lon": -122.4194},
    "City4": {"lat": 51.5074, "lon": -0.1278},
    "City5": {"lat": 48.8566, "lon": 2.3522},
    "City6": {"lat": 35.6895, "lon": 139.6917},
    "City7": {"lat": -33.8688, "lon": 151.2093},
    "City8": {"lat": 55.7558, "lon": 37.6173},
    "City9": {"lat": -23.5505, "lon": -46.6333},
    "City10": {"lat": 19.4326, "lon": -99.1332}
}

# Create a map centered around an average location
map = folium.Map(location=[40, -20], zoom_start=2)

# Add markers for each city
for city, coords in cities.items():
    folium.Marker([coords['lat'], coords['lon']], popup=city).add_to(map)

# Display the map
map
```
